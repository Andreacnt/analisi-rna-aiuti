# RNA — Aiuti di Stato alle imprese italiane

**Fonte:** Registro Nazionale Aiuti di Stato (MIMIT)  
**Query:** DuckDB su DataCivicLab  
**Periodo:** 2017–2026  
**Righe totali:** ~17 milioni

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
```

```python
trend = pd.read_csv('trend_annuale.csv')
procedimenti = pd.read_csv('procedimenti.csv')

trend.head()
```

```python
fig, ax = plt.subplots()
bars = ax.bar(trend['anno'].astype(str), trend['totale'] / 1e9, color='#1976d2')
ax.set_title('Aiuti di Stato alle imprese italiane (miliardi € per anno)')
ax.set_xlabel('Anno')
ax.set_ylabel('Miliardi €')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('trend_annuale.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
procedimenti_clean = procedimenti.dropna(subset=['procedimento'])

fig, ax = plt.subplots()
bars = ax.barh(procedimenti_clean['procedimento'], procedimenti_clean['totale'] / 1e9, color='#d32f2f')
ax.set_title('Totale aiuti per tipo procedimento (miliardi €)')
ax.set_xlabel('Miliardi €')
ax.invert_yaxis()
plt.tight_layout()
plt.savefig('procedimenti.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Insight
- Il 2020 ha avuto un picco di aiuti (COVID) — oltre 95 miliardi €
- La maggior parte degli aiuti è per Notifica (€397 mld), poi De Minimis (€204 mld)
- Il dataset copre 17 milioni di singoli aiuti alle imprese italiane
